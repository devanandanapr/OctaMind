```python
import google.generativeai as genai
import ipywidgets as widgets
from IPython.display import display, Markdown

# Configure API
API_KEY = "AIzaSyCfIaNGYEQ6I2SfVYOTTRiqz5N4HielEnU"
genai.configure(api_key=API_KEY)
model = genai.GenerativeModel(model_name="gemini-1.5-flash")

# Input fields for student profile
age_input = widgets.IntText(description='Age:', layout=widgets.Layout(width='400px'))
gender_input = widgets.Dropdown(
    description='Gender:',
    options=['Male', 'Female', 'Other'],
    layout=widgets.Layout(width='400px')
)
education_input = widgets.Dropdown(
    description='Education:',
    options=['10th', '12th', 'Diploma', 'Undergraduate', 'Postgraduate'],
    layout=widgets.Layout(width='400px')
)
income_input = widgets.Text(description='Annual Income:', layout=widgets.Layout(width='400px'))
caste_input = widgets.Dropdown(
    description='Caste:',
    options=['General', 'OBC', 'SC', 'ST', 'EWS'],
    layout=widgets.Layout(width='400px')
)
state_input = widgets.Text(description='State:', layout=widgets.Layout(width='400px'))
disability_input = widgets.Dropdown(
    description='Disability:',
    options=['No', 'Yes'],
    layout=widgets.Layout(width='400px')
)
skills_input = widgets.Text(description='Skills/Interest:', layout=widgets.Layout(width='400px'))

generate_button = widgets.Button(description='Find Schemes', layout=widgets.Layout(width='400px'))
output = widgets.Output()

# Generate recommendations
def generate_schemes(btn):
    output.clear_output()
    with output:
        prompt = f"""
        You are a smart assistant helping students in India find suitable government schemes and scholarships.
        Based on the following user details, recommend the most relevant and personalized central or state government schemes with a short explanation each:

        - Age: {age_input.value}
        - Gender: {gender_input.value}
        - Education: {education_input.value}
        - Annual Income: {income_input.value}
        - Caste Category: {caste_input.value}
        - State: {state_input.value}
        - Disability: {disability_input.value}
        - Skill Interests: {skills_input.value}

        List 3–5 best scheme recommendations with names and short descriptions.
        """
        try:
            response = model.generate_content(prompt)
            result = response.text.strip()
            display(Markdown(f"### Recommended Schemes:\n\n{result}"))
        except Exception as e:
            display(Markdown(f"**Error:** {e}"))

generate_button.on_click(generate_schemes)

# UI layout
form = widgets.VBox([
    widgets.HTML(value="<h3>Student Scheme Recommender (India)</h3>"),
    age_input,
    gender_input,
    education_input,
    income_input,
    caste_input,
    state_input,
    disability_input,
    skills_input,
    generate_button,
    output
])

display(form)
```
